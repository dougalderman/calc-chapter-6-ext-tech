```python
h = var('h')
```

```python
k = var('k')
```

```python
lam = var('lam')
```

```python
def t(h, k): return h/10 + (1/2)*k
```

```python
def g(h, k): return (1 + h/12)*k^2 - 40000
```

```python
def T(h, k, lam): return t(h, k) - lam*(g(h, k))
```

```python
plot3d(t,(h,0,300),(k,0,100),color='blue',opacity=0.8)
```

```python
plot3d(g,(h,0,300),(k,0,100),color='red',opacity=0.8)
```

```python
Th = diff(T(h, k, lam), h, 1)
```

```python
Th
```

```python
Tk = diff(T(h, k, lam), k, 1)
```

```python
Tk
```

```python
Tlam = diff(T(h, k, lam), lam, 1)
```

```python
Tlam
```

```python
solve([Th==0, Tk==0, Tlam==0], h, k, lam)
```
